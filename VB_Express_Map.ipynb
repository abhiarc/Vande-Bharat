```python
import pandas as pd
import numpy as py
from d3blocks import D3Blocks


# importing geopy library
from geopy.geocoders import Nominatim

import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
# get the response in the form of html
wikiurl="https://en.wikipedia.org/wiki/Vande_Bharat_Express"
table_class="wikitable"
response=requests.get(wikiurl)
print(response.status_code) #200 means its okay to scrape
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
rawdata=soup.find_all('table',{'class':"wikitable sortable"})[-1]
```

```python
df=pd.read_html(str(rawdata))
# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head())
```

```python
df.info()
```

```python
df.columns = ["Seq", "Start_station", "End_station", "Train_name", "Train_number", "Train_version", "Frequency", "Distnace", "Travel_time", "Speed_max", "Speed_avg", "Inauguration", "Halts"]

df.head()
```

```python
df['Distance_KM'] = df['Distnace'].str[:3]
df.head()
```

```python
df_chord = df[['Start_station','End_station','Distance_KM']]
df_chord.rename(columns = {'Start_station':'source', 'End_station':'target',
                              'Distance_KM':'weight'}, inplace = True)
df_chord
```
